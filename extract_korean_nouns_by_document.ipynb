```python
from konlpy.tag import Okt

okt = Okt()
n = 0
result = []
```

```python
# for wiki_data.txt
file = open('dataset/wiki_data.txt', encoding="utf8")

temp = []

while True:
    line = file.readline() # read the file line by line
    if not line:
        break # break if there is no more line
    
    tokenlist = okt.pos(line, stem=True, norm=True) # tokenize the line using okt
    for word in tokenlist:
        if word[1] in ["Noun"]:
            temp.append((word[0])) # save to 'temp' only if a word is a noun
    
    # when a document is over
    if line.find('</doc>'):
        if temp:
            result.append(temp) # save to 'result'
            temp = [] # 'temp' initialization
        
    n += 1
    if n % 5000 == 0:
        print(n) # to check

file.close()
```

```python
# for namu_data.txt
file = open('dataset/namu_data.txt', encoding="utf8", errors="ignore")

temp = []

while True:
    line = file.readline() # read the file line by line
    if not line:
        break # break if there is no more line
        
    # when a document is over
    if line.find('<head>'):
        if temp:
            result.append(temp) # save to 'result'
            temp = [] # 'temp' initialization
    
    tokenlist = okt.pos(line, stem=True, norm=True) # tokenize the line using okt
    for word in tokenlist:
        if word[1] in ["Noun"]:
            temp.append((word[0])) # save to 'temp' only if a word is a noun
        
    n += 1
    if n % 5000 == 0:
        print(n) # to check

file.close()
```
